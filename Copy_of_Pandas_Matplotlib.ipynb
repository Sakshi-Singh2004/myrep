# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
x
```

```python
x_sq = x**2
x_sq
```

```python
plt.plot(x, x_sq, marker="o", label="y=x^2")
plt.plot(x, x, marker="*", color='red', label="y=x")
plt.xlabel("x")
plt.ylabel("Graph")
plt.title("Comparison func")
plt.legend() # it displays the labels.
plt.show()
```

```python
plt.plot(x, x, marker="*", c='red')
plt.show()
```

## Scatter Plot

```python
plt.scatter(x, x_sq, marker="o", label="y=x^2")
plt.scatter(x, x, marker="*", color='red', label="y=x")
plt.xlabel("x")
plt.ylabel("Graph")
plt.title("Comparison func")
plt.legend() # it displays the labels.
plt.show()
```

# Bar Graph

```python
x = np.array([1,2,3,4,5])

sales_in_2018 = np.array([5, 3, 1, 8, 9 ])
sales_in_2019 = np.array([2, 6, 9, 6, 7])
```

```python
plt.bar(x, sales_in_2018, width=0.3)
plt.bar(x+0.3, sales_in_2019, width=0.3, alpha= 0.5)
plt.xlabel("Car number")
plt.ylabel("Sales")
plt.show()
```

- OpenCV (Open Source Computer Vision Library) is a library of programming functions mainly for real-time computer vision. Originally developed by Intel,

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('group.jpeg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
```

```python
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
all_faces = detector.detectMultiScale(img)
all_faces
```

```python
for one_face in all_faces:
    x, y, w, h = one_face
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,0), 2)
```

```python
plt.imshow(img)
plt.show()
```
